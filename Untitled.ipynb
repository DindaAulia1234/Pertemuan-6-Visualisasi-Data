```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat figure dengan ukuran tertentu
plt.figure(figsize=(10, 6))

# Menghitung histogram dan PDF
counts, bin_edges = np.histogram(cancer_sur['Age'], bins=20, density=True)
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

# Membuat PDF plot (garis biru)
plt.plot(bin_centers, counts, color='#1f77b4', linewidth=2)

# Menghitung dan membuat CDF plot (garis oranye)
cdf = np.cumsum(counts * np.diff(bin_edges))
plt.plot(bin_centers, cdf, color='#ff7f0e', linewidth=2)

# Menambahkan garis vertikal pada umur 70
plt.axvline(x=70, color='black', linestyle='-', linewidth=1)

# Mengatur batasan sumbu
plt.xlim(35, 82)
plt.ylim(0, 1.1)

# Menambahkan label
plt.xlabel('Age')

# Menghilangkan grid
plt.grid(False)

plt.show()
```
